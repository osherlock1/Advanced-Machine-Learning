```python
%%writefile forward_pass_exercise.py
from manim import *

class ForwardPassExercise(Scene):
    def construct(self):
        input_node = Dot(LEFT * 3)
        hidden_node = Dot()
        output_node = Dot(RIGHT * 3)

        edge1 = Line(input_node.get_center(), hidden_node.get_center())
        edge2 = Line(hidden_node.get_center(), output_node.get_center())

        self.add(input_node, hidden_node, output_node, edge1, edge2)

        input_label = MathTex("x = 0.8").next_to(input_node, LEFT)
        hidden_label = MathTex("h = x \\cdot w + b").next_to(hidden_node, UP)
        output_label = MathTex("\\hat{y} = \\sigma(h)").next_to(output_node, RIGHT)

        self.play(Write(input_label), Write(hidden_label), Write(output_label))

        input_val = MathTex("0.8").move_to(input_node.get_center() + DOWN * 0.5)
        self.play(FadeIn(input_val))
        self.play(input_val.animate.move_to(hidden_node.get_center()), run_time=1)

        hidden_val = MathTex("1.2").next_to(hidden_node, DOWN)
        self.play(FadeIn(hidden_val))

        output_val = MathTex("0.73").next_to(output_node, DOWN)
        self.play(FadeIn(output_val))

        self.wait(2)
```

```python
!manim -pql forward_pass_exercise.py ForwardPassExercise
```
